```python
# Read jsonl file and convert to pandas dataframe
import json

with open('data/graphs-2023-02-01.jsonl') as f:
    data = [json.loads(line) for line in f]
```

```python
import pandas as pd
df = pd.DataFrame(data)
```

```python
# Extract the nodes and edges from the dataframe
df = df[['nodes']]
df = df.explode('nodes')
```

```python
# Extract the dictionary in the nodes to columns
df_a = pd.DataFrame(df['nodes'].tolist())
```
